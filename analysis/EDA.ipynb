Perform multivariate analysis.
Results: 
- The correlation between the gr liv area and the sale price is 0.7086.

- The correlation between the total bsmt sf and the sale price is 0.6136.
- The correlation between the year built and the sale price is 0.3678.
Summary: 
- The features that have the highest correlation with the sale price are the overall qual and the gr liv area.
- The features that have the lowest correlation with the sale price are the year built and the total bsmt sf.
Interpretation: 
- The features that have the highest correlation with the sale price are the overall qual and the gr liv area. This means that the overall quality of the house and the living area of the house have a strong positive correlation with the sale price.
- The features that have the lowest correlation with the sale price are the year built and the total bsmt sf. This means that the year the house was built and the total square footage of the basement have a weak positive correlation with the sale price.
